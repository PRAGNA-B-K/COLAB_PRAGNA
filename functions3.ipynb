```python
def mystery(lst):
    seen = []
    for x in lst:
        if x not in seen:
            seen.append(x)
    return seen
my_list=[1,2,2,3]
mystery(my_list)
```

```python
my_list = [1,2,3]
def double_pairs(lst):
    return[(x, x*2) for x in lst if x%2==1]
double_pairs(my_list)
```

```python
scans = [[1,0,1], [0,1,0], [1,1,0]]  # pest presence
alerts = []
for i in range(len(scans)):
    row_count = 0
    for j in range(len(scans[i])):
        if scans[i][j] == 1:
            row_count += 1
    if row_count > 1:
        alerts.append(f"Row{i} quarantine")
print(alerts)
```

```python
nums=[1,2,3,4]
for n in nums:
  if n%2==0:
    continue
  print(n)
```

```python
def analyze(lists):
    empty_count = 0
    i = 0
    while i < len(lists):
        if not lists[i]:
            empty_count += 1
        i += 1
    return empty_count
print(analyze([[], [1], [2,3]]))
```

```python
fields=[[10,0,15],[0,20,0],[5,10,25]]
irrigated=0
for row in range(len(fields)):
    for col in range(len(fields[row])):
        if fields[row][col] < 12:
            irrigated += 1
            continue
        if fields[row][col]>20:
            break
print(irrigated)
```
