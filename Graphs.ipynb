```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([10,20,30,40,50,60])
labels=['india','america','europe','asia','africa','australia']
colors=['red','yellow','black','hotpink','green','indigo']
plt.pie(x,colors=colors,labels=labels,autopct='%1.1f%%',wedgeprops={'edgecolor':'blue','linewidth':2})
plt.axis('equal')
plt.show()
```
